```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
imputer = SimpleImputer(strategy='mean')
X_train_transformed = imputer.fit_transform(X_train)
X_test_transformed = imputer.fit_transform(X_test)
```

```python
reg = LinearRegression()
reg.fit(X_train_transformed,y_train)
```

```python
y_pred = reg.predict(X_test_transformed)
```

```python
print('the r2 score metrics of this method is',r2_score(y_test,y_pred))
```

```python
print('the mean absolute error of this method is',mean_absolute_error(y_test,y_pred))
```
